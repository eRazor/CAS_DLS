- Sort the array
- Add comments to your code


````python
import numpy as np

x = np.array([10, 17, 13, 21, 14, 15])
y = np.array([12, 26, 10])

z=np.concatenate((x,y), axis=0)
z_combined=np.mean(list(x) + list(y)) #testing list combine
z_mean=np.mean(z)
z_med=np.median(z)
z_sorted=np.sort(z)

print(f"z_combined:{z_combined} z_mean:{z_mean} z_med:{z_med} z_sorted:{z_sorted}")

````

🌶️<font color='red'>TODO 6</font>: Write a function course_attendance(registered_students, number_of_participants)

The function returns a string:
- "nobody showed up" (if number_of_participants is zero)
- "some students did not show up" (if number of participants is less than number of registered students)
- "full house today" (if the number of participants is equal to the number of registered students)



````python
# write your function code here, then test it
def course_attendance(registered_students, number_of_participants) -> str:
    if number_of_participants == 0:
        print("nobody showed up")
    elif number_of_participants<registered_students:
        print("some students did not show up")
    elif number_of_participants==registered_students:
        print("full house today")
    else:
        print("error")

course_attendance(1,3)
````
