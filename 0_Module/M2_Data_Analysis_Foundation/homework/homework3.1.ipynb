````python
# your code here
#calculate the z-scores
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

df['GrLivArea_zscore'] = (df['GrLivArea'] - df['GrLivArea'].mean()) / df['GrLivArea'].std()
print(df['GrLivArea_zscore'].describe())

df['GrLivArea_zscore_SciPy'] = stats.zscore(df['GrLivArea'])
print(df['GrLivArea_zscore_SciPy'].describe())

fig1= sns.boxplot(x=df['GrLivArea_zscore_SciPy'])

fig2= sns.boxplot(x=df['GrLivArea_zscore'])
fig1.plot()
fig2.plot()

mask = ((df['GrLivArea_zscore'] > 3) | (df['GrLivArea_zscore'] < -3))
df_outliers = df[mask]
df_outliers
````
