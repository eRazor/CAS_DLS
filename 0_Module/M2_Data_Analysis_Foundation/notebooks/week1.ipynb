## Biotech Constants

- Avogadro's Number: $N_A = 6.022 \times 10^{23} \text{ mol}^{-1}$

- PCR Efficiency: $E_{PCR} \in [1.8, 2.0]$

## Pharma Constants

- Therapeutic Index: $TI = \frac{LD_{50}}{ED_{50}}$

- Drug Half-life: $t_{1/2} = \frac{\ln(2)}{k_e}$

- Bioavailability: $F = \frac{AUC_{oral}}{AUC_{IV}} \times \frac{Dose_{IV}}{Dose_{oral}}$

- Volume of Distribution: $V_d = \frac{Dose}{C_0}$

- Clearance Rate: $CL = \frac{k_e \times V_d}{F}$

## Meteorology Constants

- Stefan-Boltzmann Constant: $\sigma = 5.67 \times 10^{-8} \text{ W}\cdot\text{m}^{-2}\cdot\text{K}^{-4}$
- Gas Constant for Dry Air: $R_d = 287.05 \text{ J}\cdot\text{kg}^{-1}\cdot\text{K}^{-1}$
- Standard Atmospheric Pressure: $P_0 = 101,325 \text{ Pa}$
- Earth's Average Albedo: $\alpha_{Earth} = 0.3$
- Specific Heat of Dry Air: $c_p = 1005 \text{ J}\cdot\text{kg}^{-1}\cdot\text{K}^{-1}$
- Adiabatic Lapse Rate: $\Gamma_d = 9.8^{\circ}\text{C}/\text{km}$

## Nutrition Science Constants

- Energy in Carbohydrates: $E_{carb} = 4 \text{ kcal/g}$
- Energy in Protein: $E_{protein} = 4 \text{ kcal/g}$

- BMI Calculation: $BMI = \frac{\text{weight (kg)}}{\text{height (m)}^2}$

## Hurricane Constants

### Saffir-Simpson Hurricane Wind Scale
- Category 1: $v_{wind} \in [74, 95] \text{ mph } ([119, 153] \text{ km/h})$

- Category 4: $v_{wind} \in [130, 156] \text{ mph } ([209, 251] \text{ km/h})$

### Pressure Measurements
- Standard atmospheric pressure: $P_0 = 1013.25 \text{ hPa}$

## Creating Functions

Functions are reusable blocks of code. Let's look at how to define a function that calculates electric power using Watt's Law:

$P = U \times I$

Where:
- P is power in Watts
- U is voltage in Volts
- I is current in Amperes

````python
def electric_power(U, I):
    """Calculates the electric power

    Parameters
    ----------
    U : in Volt
    I : in Amp

    Returns
    -------
    Int
        die tatsächliche elektrische Leistung P is das Produkt der elektrischen Spannung U und der elektrischen Stromstärke
    """
    
    return U * I
````

Let's test our function with different values:

````python
# Test the function with different values
print(f"Power with 220V and 20A: {electric_power(220, 20)} Watts")
print(f"Power with 110V and 5A: {electric_power(110, 5)} Watts")
````

## Importing Packages in Python

Python has a built-in `statistics` module for basic statistical operations. Let's use it to calculate the average age from our dictionary:

## Data Analysis Libraries

### NumPy

NumPy (Numerical Python) is the foundation for scientific computing in Python. It provides:
- Multi-dimensional arrays
- Mathematical functions
- Tools for linear algebra, Fourier transform, and random number generation

````python
!pip install numpy
# Import NumPy with the standard alias 'np'
import numpy as np

# Create a 2D array and perform operations
x = np.arange(15, dtype=np.int64).reshape(3, 5)
x[1:, ::2] = -99  # Set every second element in rows 1 and 2 to -99
print("Array x:")
print(x)

# Find maximum value in each row
print("\nMaximum value in each row:")
print(x.max(axis=1))
````

````python
# Generate random numbers
rng = np.random.default_rng(42)  # 42 is a seed for reproducibility
samples = rng.normal(size=1000)  # 1000 samples from normal distribution

print(f"Mean: {samples.mean():.2f}")
print(f"Standard deviation: {samples.std():.2f}")
print(f"Min: {samples.min():.2f}, Max: {samples.max():.2f}")
````

### Matplotlib and Seaborn

Visualization is crucial for data analysis. Python offers two powerful libraries:

- **Matplotlib**: The foundation for most Python plotting
- **Seaborn**: Built on Matplotlib, provides statistical visualizations with a higher-level interface

Let's see a quick example:

````python
# Import visualization libraries
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
# Set the visual theme for the plots
# - style="whitegrid" adds a subtle grid on a white background for better readability
# - palette="muted" uses a set of subdued colors that work well together and are easy on the eyes
sns.set_theme(style="whitegrid", palette="muted")

# Create a new matplotlib figure with specific dimensions
# - figsize=(10, 5) sets the width to 10 inches and height to 5 inches
# - This creates a rectangular plot that's good for visualizing distributions
plt.figure(figsize=(10, 5))

# Generate a histogram of the 'samples' data array
# - histplot() creates a histogram showing the frequency distribution of values
# - kde=True adds a Kernel Density Estimate curve, which smoothly approximates 
#   the probability density function of the underlying distribution
sns.histplot(samples, kde=True)

# Add a descriptive title to the plot
# - This informs viewers that we're looking at a normal distribution
plt.title("Normal Distribution")

# Label the x-axis to indicate what the values represent
# - In this case, the x-axis shows the actual values from our samples
plt.xlabel("Value")

# Label the y-axis to indicate what's being counted
# - The y-axis represents the frequency or count of occurrences for each value range
plt.ylabel("Count")

# Display the completed visualization
# - This command renders the plot and shows it in the output
# - In Jupyter notebooks, this displays the plot inline
plt.show()
````

## Exploring Real Data

The seaborn package also contains many datasets.

Check the notebook ```seaborn_datasets.ipynb``` in the demo folder of your repository to see all the available datasets as well as many visualisations.

Let's look at a real dataset: Palmer Penguins. This dataset contains measurements of penguins in Antarctica.

````python
# Load the penguin dataset
penguins = sns.load_dataset("penguins")

# Look at the first few rows
print("First 5 rows:")
penguins.head()
````

````python
# Get basic statistics
print("Summary statistics:")
penguins.describe()
````

## Creating a Visualization

Let's create a visualization showing the relationship between different penguin measurements, colored by species.

````python
# Create a swarm plot to show body mass distribution by sex and species
plt.figure(figsize=(12, 6))
sns.swarmplot(data=penguins, x="body_mass_g", y="sex", hue="species")
plt.title("Penguin Body Mass by Sex and Species")
plt.xlabel("Body Mass (g)")
plt.show()
````

Let's create another visualization showing the relationship between bill length and bill depth for different species:

````python
# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", 
                hue="species", style="species", s=100, alpha=0.8)
plt.title("Bill Dimensions by Penguin Species")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.show()
````
