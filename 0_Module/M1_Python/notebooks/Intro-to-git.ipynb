# Source Control and Versioning (Beginner Friendly)

Welcome! In this notebook, you will learn what source control is and how to use Git.

This is for beginners taking their first Python course. No prior Git knowledge needed.

## What is Source Control?

Source control (also called version control) is a way to save your code with history.

Think of it like a time machine for your files:
- You can save a snapshot of your work (called a commit).
- You can go back if you break something.
- You can see what changed and who changed it.
- You can work with others without overwriting each other.

## What is Git?

Git is a popular tool for source control. It runs on your computer.
You can also connect it to an online service like GitHub, GitLab, or Bitbucket to back up and share your code.

Key words:
- Repository (repo): the folder tracked by Git.
- Commit: a saved snapshot of your changes with a message.
- Branch: a line of work (like a copy) where you can make changes safely.
- Remote: the online copy (e.g., on GitHub).

## Install Git (quick note)

- Windows: install from https://git-scm.com and accept defaults.
- macOS: install Xcode Command Line Tools or use Homebrew: brew install git.
- Linux (Debian/Ubuntu): sudo apt-get install git.

After installing, open a terminal (or the Jupyter terminal) and run: git --version.
You should see a version number, e.g., git version 2.x.x.

## First-time Git Setup (do once per computer)

Tell Git your name and email (they will appear on your commits). Use your own name and email:


git config --global user.name 'Your Name'
git config --global user.email 'your@email.com'


Optional: make Git output easier to read and set your default branch name:


git config --global init.defaultBranch main
git config --global color.ui auto


## Create a New Repository (start tracking a folder)

Steps:
1. Create a new folder for your project.
2. Go into that folder in your terminal.
3. Run git init to start a Git repository.

Example commands:


mkdir my-first-project
cd my-first-project
git init


You will now have a hidden .git folder. This means Git is tracking this folder.

## Add a File and Make Your First Commit

1. Create a simple Python file named hello.py with one line:


print('Hello, world!')


2. Ask Git what is going on:


git status


You should see hello.py as an untracked file (not saved yet).

3. Tell Git to track the file (stage it):


git add hello.py


4. Save a snapshot (commit) with a short message:


git commit -m 'Add hello.py'


Now your work is saved in Git history. If you change hello.py and commit again, you will have a new snapshot.

## See What Changed

- git status shows what files changed and what is staged.
- git diff shows the exact line changes (before staging).
- git add stages changes to be included in the next commit.
- git diff --staged shows what will be committed.
- git log shows the history of commits.

Try this:
1. Edit hello.py to say: print('Hello, biology!')
2. Run git status and git diff.
3. Stage and commit your change:


git add hello.py
git commit -m 'Update greeting to biology'


4. See your history:


git log --oneline


## Ignore Files You Do Not Want to Track

Some files should not be in Git (e.g., temporary files, data outputs). Create a .gitignore file to list them.

Example .gitignore content:


__pycache__/
*.pyc
.DS_Store
data/


Add and commit it:


git add .gitignore
git commit -m 'Add .gitignore'


## Create and Use a Branch (safe place to try changes)

A branch lets you work on a feature or experiment without touching the main code.

Commands:


git branch feature-readme       # create a branch
git checkout feature-readme     # switch to it


Or do both at once:


git checkout -b feature-readme


Now edit files and commit as usual. When done, merge back to main:


git checkout main
git merge feature-readme


If there are conflicts, Git will ask you to fix them. For beginners, keep changes small and clear to avoid conflicts.

## Connect to GitHub (optional but recommended)

A remote is an online copy of your repo.

Steps (summary):
1. Create a free account at https://github.com.
2. Click New repository. Name it my-first-project (no spaces).
3. Do not add a README yet (or if you do, follow GitHub instructions to pull first).
4. In your terminal, link your local repo to GitHub:


git remote add origin https://github.com/your-username/my-first-project.git
git branch -M main
git push -u origin main


Next time, just do:


git push


To get changes from GitHub to your computer:


git pull


## Common Git Commands (cheat sheet)

- git init start a repo in the current folder
- git status see what changed
- git add <file> stage a file
- git commit -m 'message' save a snapshot
- git log --oneline see history
- git diff see line-by-line changes
- git checkout -b <branch> create and switch to a branch
- git checkout <branch> switch branches
- git merge <branch> merge a branch into the current one
- git remote -v see remotes
- git push upload to remote (e.g., GitHub)
- git pull download and merge from remote

## Tips for Success

- Commit small and often with clear messages (what you changed and why).
- One idea per commit helps you read history later.
- Use branches for new features or experiments.
- Add a README.md to explain your project and how to run it.
- Use .gitignore to keep junk files out of your repo.
