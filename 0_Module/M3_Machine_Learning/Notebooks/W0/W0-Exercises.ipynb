## 1.1
Write a function that gets your birth year and return your age, for simplicity assume that you were born 1.1.

hint: you can use classic function inputs f(curernt_year, birth_year) or use input() method to enter the year of birth

```python
# your code
import datetime as dt

def print_current_age(year, month,day):
    now = dt.datetime.now()
    age = now- dt.datetime(year, month, day)
    print("Current your age is:", age.days // 365, "years")


input_birthday=input("Enter your birthday (YYYY-MM-DD): ")
year, month, day = map(int, input_birthday.split('-'))
print_current_age(year, month, day)
```

Write a function which will find all numbers divisible by 7 but are not a multiple of 5, between given range (both border limits of range included)

hint: list method append()

```python
# your code
def process_list(input_list):
    result=[]
    for element in input_list:
        #print(element)
        if (element%7 == 0) & (element%5!=0):
            result.append(element)
    return result

print(process_list(range(1,36)))
```

## 5.3
Write a function to reverse an order of words in my_string

Hint: you can use joint(), and split() methods of string

```python
# your code
def reverseWords(somestring):
    wordlist =[]
    for word in somestring.split()[::-1]:
        wordlist.append(word)
    return " ".join(wordlist)

print(reverseWords("This is a test."))
```

### 8.1
Write a function to form a frequency table of the letters in the string, that is, how many times each letter appears

Output for word Mississippi => {'M': 1, 'i': 4, 's': 4, 'p': 2}

hint:
- you can itterate through your string and create dictionary for each letter in your string
- use dictionary method get()

```python
# your code

def LetterFreq(somestring):
    sLetters=set(somestring)
    dLetter={}
    for element in list(sLetters):
        dLetter[element] =somestring.count(element)
    return dLetter

dFreq = LetterFreq("Hello World!")
print(dFreq)
```

### 8.2 
Sort the results of previous function in alphabetical order
hint: usable dictionary method items(), and list methods sort()

```python
# your code
sorted_dict = dict(sorted(dFreq.items(), key=lambda item: item[1])) 
print(sorted_dict)
```

## 10 
Write a function to ask for anumner and to check if it is a prime and print corresponding result.

Hint: Prime is divisible only by 1 and itself.

```python
# your code
def is_prime(n):
    """Check if a number is prime."""
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    
    # Check divisibility by numbers of the form 6k±1
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    
    return True

print(is_prime(67))
```
