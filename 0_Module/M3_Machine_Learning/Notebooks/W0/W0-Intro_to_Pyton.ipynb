### Identity operators
- is/is not
- Used to check if two values (or variables) are located on the same part of the memory
- Two variables that are equal does not imply that they are identical

```python
x1 = 5
y1 = 5
x2 = 'Hello'
y2 = 'Hello'
x3 = [1,2,3]
y3 = [1,2,3]

print(x1 is not y1)
# x1 and y1 are integers of the same values, so they are equal as well as identical

print(x2 is y2)
# x2 and y2 are the same strings, thus equal and identical.

# Output: False
print(x3 is y3)
# x3 and y3 are lists. Equal but not identical.
# It is because the interpreter locates them separately in memory although they are equal.
```

### Bitwise operators
- Act on operands as if they were strings of binary digits
- Operate bit by bit

```python
def decimalToBinary(n):
    return bin(n).replace("0b", "")

x = 10
print("{0:#b}".format(x))
print (bin(x))
print(decimalToBinary(x))

y = 2
print("{0:#b}".format(y))
print (bin(y))
print(decimalToBinary(y))

# bitwise AND
print('x & y is',x&y)
# bitwise OR
print('x | y is',x|y)
# bitwise NOT
print('~x is',~x)
# bitwise XOR
print('x ^ y is',x^y)
#logical shifts - useful to perform multiplication or division of unsigned integers by powers of two
# bitwise right shift
print('x >> y is',x>>y)
# bitwise left shift
print('x << y is',x<<y)
```

```python
#Python allows to use variable-length arguments with * operator
def sumnnumbers(a,*b):
    # forst we set the variable c to be 0, because later we will assing a nex value in each step of for loop
    c = 0
    for number in b:
        c = c + number # this sums all number in b input
    c = a + c # add a to the sum of b
    return c
```

#### Python Example
Having a list variable. How to select arbitrary indices from a list?



```python
original_list = [14,23,32,44,57,68,79] # can be list with age of users
list_of_indices = [1,3,5] # indices with users who smokes

# first, we define the empty list
list_of_smokers_age = []
for ind in list_of_indices:
    print(original_list[ind]) # this prints out the age of smokers
    list_of_smokers_age.append(original_list[ind]) # here we use build-in method append(), which is made on data type list and adds a value to a given list
print("age of all users:", original_list)
print("age of all smokers:", list_of_smokers_age)
```

```python
# one liner with list comprehensions (this concept of list comprehensions will be explained later)
# this example shows the magic of Python ;)
list_of_smokers_age = [original_list[ind] for ind in list_of_indices]
print(list_of_smokers_age)
```

#### Tuple as return values
- Functions in basic form always only return a single value
- Making that value a tuple, we can effectively group together as many values as we like, and return them together


```python
# Example: "Function that returns (circumference (c), area (a)) of a circle of radius r "
import math # in this module, there is function pi, which we will use
def get_values(r):
    c = 2 * math.pi * r
    a = math.pi * r * r
    return (c, a)

print(get_values(1))
print(type(get_values(1)))
```

```python
# We can access each possition in tuble by iterable unpacking of tuple
# For example, we can unpack the tuple returned by function get_values and assing it to variable c and a

# get unpacked c and a values
c,a = get_values(1)
print(c)
print(a)
print(type(c), type(a) )

# get tuple
ca = get_values(1)
print(ca)
print(type(ca))
```

## Aliasing
- Alliasing happens whenever one variable is assigned to another variable
- Depends on data type


### Aliasing for Data Type String
- Not an issue because Strings are immutable - cannot be changed
- Whenever we change a string, Python actually creates a new string behind the scenes.




```python
# lets define two variables, each refer to a string object "banana"
a = "banana"
b = "banana"
# do a and b have the same value = are they equal?
print(a == b) # identical!!!

# do a and b refer to the same object = are they identical?
print(a is b) # identical!!!
```

- Since strings are immutable, Python optimizes resources by making two names that refer to the same string value refer to the same object

![title](https://drive.google.com/uc?export=view&id=1wvi6XG0u04XYCqJZ8V0w-v-TMkuHkZO0)

- No Aliasing problem

```python
# example to show the immutability of string and no aliassing for string
a = 'banana'
b = a
a = "apple"
print(a, b)
a is b
```

### Aliasing for Data Type List and Dictionary
- Lists and dictionaries - mutable - i.e. can be altered if identical
  
  ![title](https://drive.google.com/uc?export=view&id=1xBfUZlUpfA_x1NDXdAWd3bhtFecaJ7Re)

- If we assign one variable to another, both variables refer to the same object

 ![title](https://drive.google.com/uc?export=view&id=1cwa3MTVYHTj0rMK8kuwWiXsTBirJopBl)

```python
# defining two separate lists
a = [1, 2, 3]
b = [1, 2, 3]
print(a == b) # are equal
print(a is b) # are not identical!!!
```

```python
# assinging one list to another - i.e. making aliased variable
a = [1, 2, 3]
b = a
print(a == b) # are equal
print(a is b) # are identical!!!
```

```python
# Changing one implies changes in another!!! - aliased
b[0] = 5
a
```

### Avoiding Aliasing
- Cloning list

```python
a = [1, 2, 3]
b = a[:] # cloning
print(b)
b[0] = 5
print(a)
```

- Copying of dictionaries

```python
opposites = {"up": "down", "right": "wrong", "yes": "no"}
alias = opposites # alias and opposites refer to the same object
copy = opposites.copy()  # copy refers to a fresh copy of the same dictionary = shallow copy
```

```python
# if we modify alias, opposites is also changed
alias["right"] = "left"
opposites["right"]
```

```python
# if we modify copy, opposites is unchanged
copy["right"] = "privilege"
opposites["right"]
```

##### Regular parentheses usage for generator expressions
- generators
    - an advance topic (comprehensions and iterators)
    - very useful
    - allows to describe a sequence of data without creating each element of that sequence
- generator is iterable
    - can be placed inside for loop
    - can represent potentially infinite sequence of data without actuall need to install an infinite amount of RAM on our computers
    - we can retrieve items from our iterable one at a time

```python
# example of creating generator object
# generator expressions are surrounded by parenthesis ()
g = (n * n for n in range(10)) # creates generator

l = [n * n for n in range(10)] # creates list using given generator g

# compare the following
print(g)
print(list(g))
print(l)
```

```python
# example of using generator to iterate through a range of values

r = range(10) # range is a build-in function that creates a sequence of numbers in given range "range(start, stop, step)", default start and step is 0 and 1, resp.
print(type(r))
print(list(r)) # if you want to see what values are in the variable r

g = (n * n for n in range(10)) # creates iterator
for item in g: # iterrable - for each item generated by our generator we do defined steps (in this case it is only print)
    print(item)
```

```python
# Example using generator:
# Having a given list of integers, how to convert it to a string that contains given elements or separators

mylist = [10, 20, 30, 344] # define list
ch = ',' # string defined as a comma character that we want to add to our resulting string and that separates the original list values

# we use join() build-in methods, basic syntax = string.join(iterable)
s2 = ch.join(str(x) for x in mylist) # error due to type list, note that "str(x) for x in mylist" is our generator which is an iterable
print(s2)

# with
# ','.join(mylist)
# there will be an error due to type list in join() argument

# example oj join() function
' and '.join(['1','2'])

s3 = '--'.join(str(x) for x in mylist) # error due to type list, note that "str(x) for x in mylist" is our generator which is an iterable
print(s3)
```

##### Curly braces for usage in concept of "str.format"
- string formating
- you can use printf-style with % or the modern f-string
- output of f-string is not a print, you can assign it to a new variable, save it, print it

```python
# example
name = 'Martin' # define string variable
print("Hello, {0}".format(name)) # use str.format() method
```

```python
# anpther way of using str.format()
# manual vs auto numbering in curly braces
a, b, c = [1, 2, 3]
print('a = {}, b = {}, c = {}'.format(a, b, c)) # here braces {} are empty, so variables are auto assgned to positions
```

```python
a, b, c = [1, 2, 3]
print('a = {0}, b != {0}, c = {2}'.format(a, b, c)) # here braces {} contain index of variable to be assigned
```

```python
print('a = {x}, b = {y}, c = {z}'.format(x=a, y=b, z=c))
```

##### Curly braces for f-strings
- adding “f” before the opening quotes allows us to use curly braces to interpolate just about any Python expression we want inside of a string — from variable names to operations to function/method calls
- check string methods https://www.w3schools.com/python/python_ref_string.asp


```python
print(f"Hello, {name}")
```

```python
print(f"Hello, {name.upper()}")
```

```python
print(f"Hello, {name.split('a')}")
```

```python
# How to plot a simple graph
# import modules using "as" to be able to refer to them via defined shortcut
import matplotlib.pyplot as plt
import numpy as np
# Generate a sequence of numbers from -10 to 10 with 100 steps in between
x = np.linspace(-10, 10, 100)
# Create a second array using sine
y = np.sin(x)
# The plot function makes a line chart of one array against another
plt.plot(x, y, marker="x")
```

```python
plt.plot(x, y, 'o')
```

```python
fig, ax = plt.subplots() # initiate figure and axis to use them for plots
for color in ['tab:blue', 'tab:orange', 'tab:green']:
    n = 75
    x, y = np.random.rand(2, n)
    scale = 200.0 * np.random.rand(n)
    # scatter() is a method that can be used in axis object to plot a scatter plot
    ax.scatter(x, y, c=color, s=scale, label=color,
               alpha=0.3, edgecolors='none')

ax.legend()
ax.grid(True)
```

```python
r = np.linspace(0, 1, 100)
theta = np.linspace(0, 8*np.pi, 100)
x = r * np.sin(theta)
y = r * np.cos(theta)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.plot(x, y, 'C3', lw=3)
ax1.scatter(x, y, s=120)
ax1.set_title('Lines on top of dots')

ax2.plot(x, y, 'C3', lw=3)
ax2.scatter(x, y, s=120, zorder=2.5)  # move dots on top of line
ax2.set_title('Dots on top of lines')

plt.tight_layout()
```

### Matplotlib vs Seaborn

```python
# # !pip install jupyterthemes
# from jupyterthemes import jtplot
# jtplot.style(theme='default', context='notebook', ticks=True, grid=False)

import matplotlib.pyplot as plt
import pandas as pd

# Initialize Figure and Axes object
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.set_title('plot created using matplotlib')
ax2.set_title('plot created using seaborn')

# Load in data
tips = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")

# Create violinplot which draw a combination of boxplot and kernel density estimate.
ax1.violinplot(tips["total_bill"], vert=False)


import seaborn as sns
# Load the data
tips = sns.load_dataset("tips")

# Create violinplot
sns.violinplot(x = "total_bill", data=tips, ax=ax2)



```
