## 9
Write a function to create a period index representing all monthly boundaries of a given year.
Also print start and end time for each period object (i.e month) in the said index.

hint: period_range,
+ check differences between datetimeindex and periodindex of pandas time series objects
+ Each record in a datetimeindex is a pd.Timestamp and represents a single point in time. Each record in a periodindex is a pd.Period and represents a time interval. A period contains a start_time and end_time attribute. They contain a lot of similar methods/attributes, and are often used interchangeably though.

```python
# your code
def getmonths(dtstart,dtend):
    all_months = pd.period_range(pd.to_datetime(dtstart),pd.to_datetime(dtend),freq="M")
    return all_months

def getstartenddates(monthlist):
    startenddates=[]
    for e in monthlist:
        startenddates.append(e.start_time)
        startenddates.append(e.end_time)
    return startenddates

print(getstartenddates(getmonths("2026-01-01","2026-12-31")))



```
