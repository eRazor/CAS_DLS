### Encoding features
#### One-Hot Encoding
- label encoding problem
 - numeric values can be misinterpreted by algorithms as having some sort of hierarchy/order in them
- one hot encoding solves this issue
- encodes feature to m binary features - to get numeric representation of categorical feature
  - in this strategy, each category value is converted into a new column and assigned a 1 or 0 (notation for true/false) value to the column.
  - so it encodes a feature to m binary features
  - m depends on a number of categories


#### Dummy encoding
 - python one-line of code
 - this approach is more flexible because it allows encoding as many category columns as you would like and choose how to label the columns using a prefix
 - proper naming will make the rest of the analysis just a little bit easier.

#### !!! DUMMY VARIABLE TRAP !!!
- one-hot-encoding, dummy encoding with not specified syntax
 - multicolinearity issue - features are not independent, problem with interpretability of regression models, random behaviour of model

  - using m-1 feature vector to represent each categorical feature, one categorical feature is represented by vector of 0

## Feature Engineering - Time Series Data

```python
# possibly relevant packages
# import datetime
# from dateutil.parser import parse
# import pytz

# define data as timeseries
time_stamps = ['2015-03-08 10:30:00.360000+00:00', '2017-07-13 15:45:05.755000-07:00', '2012-01-20 22:30:00.254000+05:30', '2016-12-25 00:30:00.000000+10:00']
df = pd.DataFrame(time_stamps, columns=['Time'])
df
```

```python
# convert to timestamp
ts_objs = np.array([pd.Timestamp(item) for item in np.array(df.Time)])
df['TS_obj'] = ts_objs
print(ts_objs)
df
```

```python
# useful timeseries mentods to get time relevant features

# lambda function syntax >> "lambda arguments : expression"
# A lambda function is a small anonymous function.
# A lambda function can take any number of arguments, but can only have one expression.

df['Year'] = df['TS_obj'].apply(lambda d: d.year)
df['Month'] = df['TS_obj'].apply(lambda d: d.month)
df['Day'] = df['TS_obj'].apply(lambda d: d.day)
df['DayOfWeek'] = df['TS_obj'].apply(lambda d: d.dayofweek)
df['DayName'] = df['TS_obj'].apply(lambda d: d.day_name())
df['DayOfYear'] = df['TS_obj'].apply(lambda d: d.dayofyear)
df['WeekOfYear'] = df['TS_obj'].apply(lambda d: d.weekofyear)
df['Quarter'] = df['TS_obj'].apply(lambda d: d.quarter)
df[['Time', 'Year', 'Month', 'Day', 'Quarter', 'DayOfWeek', 'DayName', 'DayOfYear', 'WeekOfYear']]
```

```python
df['Hour'] = df['TS_obj'].apply(lambda d: d.hour)
df['Minute'] = df['TS_obj'].apply(lambda d: d.minute)
df['Second'] = df['TS_obj'].apply(lambda d: d.second)
df['MUsecond'] = df['TS_obj'].apply(lambda d: d.microsecond)
df['UTC_offset'] = df['TS_obj'].apply(lambda d: d.utcoffset())
df[['Time', 'Hour', 'Minute', 'Second', 'MUsecond', 'UTC_offset']]
```

```python
# apply different feature scaling methods using scikit only on one column and see differences
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
ss = StandardScaler()
mms = MinMaxScaler()
rs = RobustScaler()
Population = pd.DataFrame(df['Population'])
Population['SS - Population'] = ss.fit_transform(df[['Population']])
Population['MM - Population'] = mms.fit_transform(df[['Population']])
Population['RS - Population'] = rs.fit_transform(df[['Population']])
Population
```

```python
# min max scaler applied on entire df
df2 = pd.DataFrame(mms.fit_transform(df), columns = df.columns)
df2
```

## Feature selection

```python
# VarianceThreshold is a simple baseline approach to feature selection.
# It removes all features whose variance doesn’t meet some threshold.
# By default, it removes all zero-variance features, i.e. features that have the same value in all samples.
from sklearn.feature_selection import VarianceThreshold
vt = VarianceThreshold(threshold=0.001)
vt.fit(df2) # we need to have same scalling on features with preserved variance - setign up the threshold can be tricky
pd.DataFrame({'variance': vt.variances_, 'select_feature': vt.get_support()}, index=df.columns).T
```

### Statistical methods

```python
import numpy as np
import pandas as pd
np.set_printoptions(suppress=True)
pt = np.get_printoptions()['threshold']

from sklearn.datasets import load_breast_cancer
bc_data = load_breast_cancer()
# print(bc_data.DESCR)

```

```python
bc_features = pd.DataFrame(bc_data.data, columns=bc_data.feature_names)
bc_classes = pd.DataFrame(bc_data.target, columns=['IsMalignant'])

# build featureset and response class labels
bc_X = np.array(bc_features)
bc_y = np.array(bc_classes).T[0]
print('Feature set shape:', bc_X.shape)
print('Response class shape:', bc_y.shape)

# We can clearly see that there are a total of 30 features in this dataset
# and a total of 569 rows of observations.
```

```python
#To get some more detail into the feature names and take a peek at the data points, you can use the following code.
# np.set_printoptions(threshold=30)
print('Feature set data [shape: '+str(bc_X.shape)+']')
print(np.round(bc_X, 2), '\n')
print('Feature names:')
print(np.array(bc_features.columns), '\n')
print('Response Class label data [shape: '+str(bc_y.shape)+']')
print(bc_y, '\n')
print('Response variable name:', np.array(bc_classes.columns))
np.set_printoptions(threshold=pt)
```

### The chi-square test
 - compute statistics between each feature and the class variable (univariate tests).
 - Selecting the top K features is more than likely to remove features having a low score and consequently they are most likely to be independent
of the class variable and hence not useful in building models.

```python
from sklearn.feature_selection import chi2, SelectKBest
skb = SelectKBest(score_func=chi2, k=15)
skb.fit(bc_X, bc_y)

```

```python
# sort and see most relevant features
feature_scores = [(item, score) for item, score in zip(bc_data.feature_names, skb.scores_)]

# sort from largest to smallest using lambda expression, check sorted command syntax
sorted(feature_scores, key=lambda x: -x[1])[:15]
```

```python
# create subset of k best
select_features_kbest = skb.get_support()
print(select_features_kbest)
feature_names_kbest = bc_data.feature_names[select_features_kbest]
print('best features = ', feature_names_kbest)

# selst only columns with selected features
feature_subset_df = bc_features[feature_names_kbest]

print(feature_names_kbest)
feature_subset_df.head(10)
```

# Test the feature set - use case

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

# selected features array
bc_SX = np.array(feature_subset_df)
# print(bc_SX.shape)

# build logistic regression model
lr = LogisticRegression(max_iter=5000)

# evaluating accuracy for model built on full featureset
full_feat_acc = np.average(cross_val_score(lr, bc_X, bc_y, scoring='accuracy', cv=2))
# evaluating accuracy for model built on selected featureset
sel_feat_acc = np.average(cross_val_score(lr, bc_SX, bc_y, scoring='accuracy', cv=2))
print('Model accuracy statistics with 5-fold cross validation')
print('Model accuracy with complete feature set', bc_X.shape, ':', full_feat_acc)
print('Model accuracy with selected feature set', bc_SX.shape, ':', sel_feat_acc)
```
