Write a function that takes two predefined lists:

- list of heights: h

```python
# predefined lists
w = [20, 30, 40, 50, 60, 70, 80]
h = [120, 130, 140, 150, 160, 170, 180]
```

```python
# one possible solution

def to_df2(w,h):
  df = pd.DataFrame()
  df['w'] = w
  df['h'] = h
  return df

# another way when using arrays
def to_df(w,h):
  a = np.array([w,h])
  a = a.T
  df = pd.DataFrame(data=a, columns=['weight', 'height'])
  return df

w = [20, 30, 40, 50, 60, 70, 80]
h = [120, 130, 140, 150, 160, 170, 180]

print(to_df(w,h))
print(to_df2(w,h))
```

Write a function  to count city wise number of people from a given data set (city, name of the person).


```python
df = pd.DataFrame({
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'city': ['California', 'Los Angeles', 'California', 'California', 'California', 'Los Angeles', 'Los Angeles', 'Georgia', 'Georgia', 'Los Angeles']
    })
```

```python
# your code

def countnumbers(df, col):
  count  = df.groupby([col]).size() # first create goupby object based on column name and then get size of each group - size returns number of rows
  # now we reset index in count series, this returns dataframe
  df_count = count.reset_index(name='Number of people')
  return df_count

c = countnumbers(df, 'city')
c
```

```python
# your code
# load packages
from sklearn.datasets import load_iris, load_breast_cancer, load_diabetes, fetch_california_housing
import pandas as pd
import numpy as np

#load dataset
cancer = load_breast_cancer()
# create dataset of a cancer
df= pd.DataFrame(data= np.c_[cancer['data'], cancer['target']],columns= list(cancer['feature_names']) + ['target'])
df
#why np.c_??? - concatenation columnwise
```

```python
# Explore groupby function of pandas print mean values for seleced groups and plot distribution of data in selected groups
import matplotlib.pyplot as plt
target_groups = df.groupby(by='target')
print(target_groups)

fig,(ax1, ax2) = plt.subplots(1,2)
ax1.hist(target_groups.get_group(0)['mean radius'])
ax2.hist(target_groups.get_group(1)['mean radius'])
```

```python
# Depending on the data you selected:
# apply one-hot/dummy encoding on selected feature
# apply scaling of selected features
# select k best features based on Chi2 test

# I will take another data with categorical features
import io
import requests
# url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
s = requests.get(url).content
df = pd.read_csv(io.StringIO(s.decode('utf-8')))
print(df)

# apply one-hot/dummy encoding on selected feature
pd.get_dummies(df['smoker'])

# apply scaling of selected features
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
ss = StandardScaler()
mms = MinMaxScaler()
rs = RobustScaler()
total_bill = pd.DataFrame(df['total_bill'])
total_bill['SS - Population'] = ss.fit_transform(df[['total_bill']])
total_bill['MM - Population'] = mms.fit_transform(df[['total_bill']])
total_bill['RS - Population'] = rs.fit_transform(df[['total_bill']])
total_bill

# select k best features based on Chi2 test - I again use cancer data
X = pd.DataFrame(cancer.data, columns=cancer.feature_names)
y = pd.DataFrame(cancer.target, columns=['target'])
```

```python
from sklearn.feature_selection import chi2, SelectKBest
skb = SelectKBest(score_func=chi2, k=10)
skb.fit(X, y)

features = skb.feature_names_in_
print(features)
#get only supported features
selected_features = features[skb.get_support()]
print(selected_features)

scores = skb.scores_
print(scores)
selected_scores = scores[skb.get_support()]
print(selected_scores)
```
