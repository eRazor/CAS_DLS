5. When You SHOULD Use Log Transform

✅ Use it when:

Data grows exponentially

Data is strictly positive

Variance increases over time

You care about relative (percentage) changes

❌ Avoid it when:

Data contains zeros or negatives (unless you use log1p)

Data is already stationary and linear

6. When Should You Use Box–Cox?
The Box–Cox transform automatically reshapes positive data to be more normal and stable, improving the performance of many machine learning models.
Use it when:
✅ Your feature is positive
✅ It is highly skewed
✅ Your model assumes normality
✅ You see fan-shaped residual plots

Avoid it when:
❌ You use tree-based models (Random Forest, XGBoost don’t need it)
❌ Your feature has zeros or negatives
