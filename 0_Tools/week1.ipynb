## 2. Variables

Variables are used to store data. In Python, you don't need to declare the type of a variable.

```python
# Assigning values to variables
name = "Alice"
age = 30
height = 1.75
is_student = True

# Printing variables
print("Name:", name)
print("Age:", age)
print("Height:", height, "meters")
print("Is student?", is_student)
```

## 3. Data Types

Python has several built-in data types. Let's explore the most common ones:

```python
# Integer (whole numbers)
age = 30
year = 2025
negative_num = -5

# Float (decimal numbers)
height = 1.75
temperature = -2.5
pi = 3.14159

# String (text)
name = "Bob"
message = 'Hello there!'
multi_line = """This is a
multi-line
string."""

# Boolean (True or False)
is_raining = False
has_passed = True

# Checking types
print("Type of age:", type(age))
print("Type of height:", type(height))
print("Type of name:", type(name))
print("Type of is_raining:", type(is_raining))
```

## 4. Basic Operations

Let's look at operations for different data types.

```python
# Numeric operations
a = 10
b = 3

print("Addition:", a + b)        # 13
print("Subtraction:", a - b)     # 7
print("Multiplication:", a * b)  # 30
print("Division:", a / b)        # 3.3333...
print("Integer division:", a // b)  # 3
print("Remainder:", a % b)       # 1
print("Exponentiation:", a ** b) # 1000

# String operations
first_name = "John"
last_name = "Doe"

# Concatenation
full_name = first_name + " " + last_name
print("Full name:", full_name)

# Repetition
repeat = "Python! " * 3
print("Repeat:", repeat)

# String length
print("Length of full_name:", len(full_name))
```

## 5. String Formatting

There are multiple ways to format strings in Python.

```python
name = "Alice"
age = 30
height = 1.75

# Method 1: Using + operator
message1 = name + " is " + str(age) + " years old and " + str(height) + " meters tall."
print("Method 1:", message1)

# Method 2: Using format() method
message2 = "{} is {} years old and {} meters tall.".format(name, age, height)
print("Method 2:", message2)

# Method 3: Using f-strings (recommended, Python 3.6+)
message3 = f"{name} is {age} years old and {height} meters tall."
print("Method 3:", message3)
```

## 6. Input and Output

Let's see how to get input from users and display output.

```python
# Getting input
name = input("Enter your name: ")
print(f"Hello, {name}!")

# Note: input() always returns a string
# To get a number, we need to convert the string
age_str = input("Enter your age: ")
age = int(age_str)  # Converting string to integer
print(f"In 5 years, you will be {age + 5} years old.")
```

## 7. Type Conversion

You can convert between different data types in Python.

```python
# String to number
num_str = "42"
num_int = int(num_str)  # Convert to integer
print(num_int, type(num_int))

num_float = float(num_str)  # Convert to float
print(num_float, type(num_float))

# Number to string
price = 19.99
price_str = str(price)  # Convert to string
print(price_str, type(price_str))

# Integer to float and vice versa
x = 10
x_float = float(x)  # Convert to float
print(x_float, type(x_float))

y = 3.14
y_int = int(y)  # Convert to integer (truncates decimal part)
print(y_int, type(y_int))

# To boolean
print("0 as boolean:", bool(0))       # False
print("1 as boolean:", bool(1))       # True
print("Empty string:", bool(""))     # False
print("Non-empty string:", bool("hello"))  # True
```
