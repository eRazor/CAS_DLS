## 1. Lists

Lists are ordered, mutable collections that can contain elements of different types.

### Creating Lists

```python
# Empty list
empty_list = []
print(f"Empty list: {empty_list}")

# List with elements of the same type
numbers = [1, 2, 3, 4, 5]
fruits = ["apple", "banana", "cherry"]
print(f"Numbers: {numbers}")
print(f"Fruits: {fruits}")

# List with elements of different types
mixed = [1, "hello", 3.14, True]
print(f"Mixed types: {mixed}")

# Nested lists
nested = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(f"Nested list: {nested}")

# Creating a list with the list() constructor
from_string = list("Python")
print(f"List from string: {from_string}")

# Creating a list with a range
from_range = list(range(1, 6))
print(f"List from range: {from_range}")
```

### Accessing List Elements

```python
fruits = ["apple", "banana", "cherry", "date", "elderberry"]

# Accessing by index (0-based)
print(f"First fruit: {fruits[0]}")
print(f"Third fruit: {fruits[2]}")

# Negative indexing (counting from the end)
print(f"Last fruit: {fruits[-1]}")
print(f"Second-to-last fruit: {fruits[-2]}")

# Slicing [start:stop:step]
print(f"First three fruits: {fruits[0:3]}")
print(f"Every other fruit: {fruits[::2]}")
print(f"Reversed list: {fruits[::-1]}")

# Accessing nested list elements
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(f"Element at row 1, column 2: {matrix[1][2]}")
```

### List Methods and Operations

```python
fruits = ["apple", "banana", "cherry"]

# Adding elements
fruits.append("date")              # Add to the end
print(f"After append: {fruits}")

fruits.insert(1, "apricot")        # Insert at specific position
print(f"After insert: {fruits}")

fruits.extend(["elderberry", "fig"])  # Add multiple items
print(f"After extend: {fruits}")

# Removing elements
fruits.remove("banana")            # Remove specific item
print(f"After remove: {fruits}")

popped = fruits.pop()              # Remove and return the last item
print(f"Popped item: {popped}")
print(f"After pop: {fruits}")

popped_index = fruits.pop(1)       # Remove and return item at index
print(f"Popped item at index 1: {popped_index}")
print(f"After pop(1): {fruits}")

# Finding elements
position = fruits.index("cherry")  # Find the index of an item
print(f"Index of 'cherry': {position}")

count = fruits.count("apple")      # Count occurrences
print(f"Count of 'apple': {count}")

# Other operations
fruits.sort()                      # Sort in place
print(f"After sort: {fruits}")

fruits.reverse()                   # Reverse in place
print(f"After reverse: {fruits}")
```

```python
# List operations
list1 = [1, 2, 3]
list2 = [4, 5, 6]


print(f"count 2 in list1: {list1.count(2)}")
# Concatenation
combined = list1 + list2
print(f"Concatenated list: {combined}")

# Repetition
repeated = list1 * 3
print(f"Repeated list: {repeated}")

# Membership testing
print(f"Is 2 in list1? {2 in list1}")
print(f"Is 7 in list1? {7 in list1}")

# Length
print(f"Length of list1: {len(list1)}")

# Min and max
print(f"Minimum in list1: {min(list1)}")
print(f"Maximum in list1: {max(list1)}")

# Sum
print(f"Sum of list1: {sum(list1)}")
```

### List Copying

```python
# Beware of shallow copying!
original = [1, 2, 3]
reference = original     # This creates a reference, not a copy

reference.append(4)      # This also modifies original
print(f"Original: {original}")
print(f"Reference: {reference}")

# Proper ways to copy
original = [1, 2, 3]

# Method 1: Slice
copy1 = original[:]

# Method 2: list() constructor
copy2 = list(original)


#%timeit

# Method 3: copy() method
copy3 = original.copy()

# Test the copies
copy1.append(4)
copy2.append(5)
copy3.append(6)

print(f"Original after modifications: {original}")
print(f"Copy1: {copy1}")
print(f"Copy2: {copy2}")
print(f"Copy3: {copy3}")
```

## 2. Dictionaries

Dictionaries are collections of key-value pairs. They are unordered (before Python 3.7), mutable, and keys must be of immutable type.

### Creating Dictionaries

```python
# Empty dictionary
empty_dict = {}
print(f"Empty dictionary: {empty_dict}")

# Dictionary with string keys
person = {
    "name": "John",
    "age": 30,
    "city": "New York",
    "is_student": False
}
print(f"Person dictionary: {person}")

# Dictionary with mixed key types
mixed_keys = {
    "name": "Product X",
    42: "The answer",
    (1, 2): "Tuple key"
}
print(f"Mixed keys: {mixed_keys}")

# Creating a dictionary with the dict() constructor
person2 = dict(name="Jane", age=25, city="Boston")
print(f"Using dict() constructor: {person2}")

# Dictionary from a list of tuples
items = [("a", 1), ("b", 2), ("c", 3)]
dict_from_tuples = dict(items)
print(f"Dict from tuples: {dict_from_tuples}")
```

### Accessing Dictionary Elements

```python
person = {
    "name": "John",
    "age": 30,
    "city": "New York",
    "email": "john@example.com"
}

# Accessing with square brackets
print(f"Name: {person['name']}")
print(f"Age: {person['age']}")

# Accessing with get() method (safer, provides default)
print(f"Email: {person.get('email')}")
print(f"Phone: {person.get('phone', 'Not available')}")

# Checking if a key exists
print(f"Does 'city' exist? {'city' in person}")
print(f"Does 'phone' exist? {'phone' in person}")

# Getting all keys, values, and items
print(f"Keys: {list(person.keys())}")
print(f"Values: {list(person.values())}")
print(f"Items: {list(person.items())}")
```

### Dictionary Methods and Operations

```python
person = {"name": "John", "age": 30, "city": "New York"}

# Adding/updating entries
person["email"] = "john@example.com"  # Add new key-value pair
person["age"] = 31                    # Update existing value
print(f"After adding/updating: {person}")

# update() method - add multiple key-value pairs
person.update({"phone": "555-1234", "country": "USA"})
print(f"After update(): {person}")

# Removing entries
removed_value = person.pop("phone")    # Remove and return a value
print(f"Removed value: {removed_value}")
print(f"After pop(): {person}")

# popitem() - remove and return a random (last in Python 3.7+) item
item = person.popitem()
print(f"Popped item: {item}")
print(f"After popitem(): {person}")

# del statement
del person["email"]
print(f"After del person['email']: {person}")

# clear() - remove all items
person_copy = person.copy()  # Make a copy first
person_copy.clear()
print(f"After clear(): {person_copy}")

# setdefault() - get a value, set default if key doesn't exist
age = person.setdefault("age", 25)      # Key exists, returns current value
print(f"Age: {age}")
height = person.setdefault("height", 175)  # Key doesn't exist, sets default
print(f"Height: {height}")
print(f"After setdefault(): {person}")
```

### Nested Dictionaries

```python
# Nested dictionary
users = {
    "user1": {
        "name": "John",
        "age": 30,
        "scores": [85, 90, 78]
    },
    "user2": {
        "name": "Jane",
        "age": 25,
        "scores": [92, 88, 95]
    }
}

# Accessing nested values
print(f"User1 name: {users['user1']['name']}")
print(f"User2 first score: {users['user2']['scores'][0]}")

# Adding to nested dictionaries
users["user1"]["email"] = "john@example.com"
users["user2"]["scores"].append(91)

print(f"Updated user1: {users['user1']}")
print(f"Updated user2 scores: {users['user2']['scores']}")
```

## 3. Sets

Sets are unordered collections of unique elements. They are mutable and elements must be of immutable type.

### Creating Sets

```python
# Empty set (can't use {} as that creates an empty dict)
empty_set = set()
print(f"Empty set: {empty_set}")

# Set with elements
fruits = {"apple", "banana", "cherry"}
print(f"Fruit set: {fruits}")

# Creating a set from a list (removes duplicates)
numbers = set([1, 2, 2, 3, 4, 4, 5])
print(f"Numbers set: {numbers}")

# Creating a set from a string (each character becomes an element)
letters = set("hello")
print(f"Letters from 'hello': {letters}")
```

### Set Operations

```python
# Set operations
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# Union (all elements from both sets)
union = set1.union(set2)
print(f"Union: {union}")
print(f"Union using |: {set1 | set2}")

# Intersection (elements in both sets)
intersection = set1.intersection(set2)
print(f"Intersection: {intersection}")
print(f"Intersection using &: {set1 & set2}")

# Difference (elements in set1 but not in set2)
difference = set1.difference(set2)
print(f"Difference (set1 - set2): {difference}")
print(f"Difference using -: {set1 - set2}")

# Symmetric difference (elements in either set, but not in both)
sym_diff = set1.symmetric_difference(set2)
print(f"Symmetric difference: {sym_diff}")
print(f"Symmetric difference using ^: {set1 ^ set2}")
```

### Set Methods

```python
fruits = {"apple", "banana", "cherry"}

# Adding elements
fruits.add("date")
print(f"After add(): {fruits}")

# Adding multiple elements
fruits.update(["elderberry", "fig", "banana"])  # Note: duplicates are ignored
print(f"After update(): {fruits}")

# Removing elements
fruits.remove("banana")    # Raises KeyError if not found
print(f"After remove(): {fruits}")

fruits.discard("kiwi")     # No error if not found
print(f"After discard() of non-existent item: {fruits}")

popped = fruits.pop()      # Remove and return an arbitrary element
print(f"Popped item: {popped}")
print(f"After pop(): {fruits}")

# Other set operations
print(f"Length of set: {len(fruits)}")
print(f"Is 'cherry' in fruits? {'cherry' in fruits}")

# Clear the set
fruits_copy = fruits.copy()
fruits_copy.clear()
print(f"After clear(): {fruits_copy}")
```

### Set Relationships

```python
# Testing set relationships
set1 = {1, 2, 3}
set2 = {1, 2, 3, 4, 5}
set3 = {6, 7, 8}

# Subset (set1 is a subset of set2 if all elements of set1 are in set2)
print(f"Is set1 a subset of set2? {set1.issubset(set2)}")
print(f"Is set1 a subset of set2 using <=? {set1 <= set2}")

# Superset (set2 is a superset of set1 if set2 contains all elements of set1)
print(f"Is set2 a superset of set1? {set2.issuperset(set1)}")
print(f"Is set2 a superset of set1 using >=? {set2 >= set1}")

# Disjoint sets (no elements in common)
print(f"Are set1 and set3 disjoint? {set1.isdisjoint(set3)}")
print(f"Are set1 and set2 disjoint? {set1.isdisjoint(set2)}")
```

## 4. Tuples

Tuples are ordered, immutable collections that can contain elements of different types.

### Creating Tuples

```python
# Empty tuple
empty_tuple = ()
print(f"Empty tuple: {empty_tuple}")

# Tuple with elements
coordinates = (10, 20)
print(f"Coordinates: {coordinates}")

# Tuple with mixed types
person = ("John", 30, "New York")
print(f"Person tuple: {person}")

# Creating a tuple with a single element (note the comma)
single_item = (42,)
print(f"Single item tuple: {single_item}")
print(f"Type of single_item: {type(single_item)}")

# Without the comma, it's not a tuple
not_a_tuple = (42)
print(f"Not a tuple: {not_a_tuple}")
print(f"Type of not_a_tuple: {type(not_a_tuple)}")

# Creating a tuple without parentheses (tuple packing)
coordinates = 10, 20
print(f"Tuple created without parentheses: {coordinates}")
print(f"Type of coordinates: {type(coordinates)}")

# Creating a tuple with the tuple() constructor
from_list = tuple([1, 2, 3])
from_string = tuple("hello")
print(f"Tuple from list: {from_list}")
print(f"Tuple from string: {from_string}")
```

### Accessing Tuple Elements

```python
person = ("John", 30, "New York", "Developer")

# Accessing by index
print(f"Name: {person[0]}")
print(f"Age: {person[1]}")

# Negative indexing
print(f"Occupation: {person[-1]}")

# Slicing
print(f"First two elements: {person[:2]}")

# Tuple unpacking
name, age, city, occupation = person
print(f"Unpacked - Name: {name}, Age: {age}, City: {city}, Occupation: {occupation}")

# Partial unpacking with *
name, *rest = person
print(f"Partial unpacking - Name: {name}, Rest: {rest}")

# Other operations
print(f"Length of tuple: {len(person)}")
print(f"Count of 'John': {person.count('John')}")
print(f"Index of 'New York': {person.index('New York')}")
```
