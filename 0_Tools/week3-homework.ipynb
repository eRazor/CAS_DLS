## Exercise 1: Basic Input and Conditional Logic (Body Temperature Check)

Scenario: In a small study, you collect a participant's body temperature (in °C). If the temperature is below 35.0, print 'Low temperature'. If it's between 35.0 and 37.5 (inclusive), print 'Normal temperature'. If it's above 37.5, print 'Fever detected'.

Instructions:
1. 🥦 Ask the user to input a temperature (as a number).
2. 🥦 Use if/elif/else to print the correct message.
3. 🌶️ Add code to handle exceptions in case the user provides an invalid input.


````python
# Your work here for Exercise 1
temperature=float(input("Please input Number:"))

if temperature<35:
    print("Low temperature")
elif temperature>=35 and temperature<=37.5:
    print("Normal temperature")
elif temperature>37.5:
    print("Fever detected")
````

## Exercise 3: Sets and Deduplication (Species Observed)

Scenario: On a short field survey, you recorded species names, but some were duplicates. Use a set to find the unique species.

Instructions:
1. 🥦 Ask the user to enter species names separated by commas (e.g., sparrow, robin, robin, hawk).
2. 🥦 Clean up extra spaces and store unique names in a set.
3. 🥦 Print the number of unique species and list them in alphabetical order.

Hint: Use str.strip(), set(), and sorted().


````python
# Your work here for Exercise 3
species=input("Please input species:").strip(" ").split(",")
print(species)
uniquespecies =[ v for k,v in enumerate(set(species))]
uniquespecies.sort()

for speciment in uniquespecies:
    print(speciment)


raw = input('Enter species names separated by commas: ')
items = [name.strip() for name in raw.split(',') if name.strip() != '']
unique_species = set(items)
print(f'Unique species count: {len(unique_species)}')
print('Species (A→Z):', ', '.join(sorted(unique_species)))
````
