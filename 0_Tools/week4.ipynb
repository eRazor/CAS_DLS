## 1. Functions in Python

Functions allow you to package code into reusable blocks. In life sciences, functions can help with:
- Calculating statistics on experimental results
- Processing and transforming biological data
- Standardizing analysis workflows

### Basic Function Definition

````python
# Basic function definition
def calculate_bmi(weight, height):
    """Calculate Body Mass Index.
    
    Args:
        weight: Weight in kilograms
        height: Height in meters
        
    Returns:
        The BMI value (weight / height^2)
    """
    bmi = weight / (height ** 2)
    return bmi

# Call the function with arguments
patient_bmi = calculate_bmi(70, 1.75)
print(f"Patient BMI: {patient_bmi:.2f}")
````

### Default Parameters and Keyword Arguments

````python
# Function with default parameters
#optional parameters is also possible
def calculate_drug_dosage(weight, age, pediatric=False, renal_impairment=False):
    """Calculate drug dosage based on patient characteristics."""
    # Base dosage calculation
    base_dosage = weight * 2
    
    # Adjust for pediatric patients
    if pediatric:
        base_dosage *= 0.75
    
    # Adjust for elderly
    if age > 65:
        base_dosage *= 0.85
    
    # Adjust for renal impairment
    if renal_impairment:
        base_dosage *= 0.6
        
    return base_dosage

# Different ways to call the function
print(f"Adult dosage: {calculate_drug_dosage(70, 45):.1f} mg")
print(f"Pediatric dosage: {calculate_drug_dosage(30, 10, pediatric=True):.1f} mg")
print(f"Elderly with renal impairment: {calculate_drug_dosage(65, 72, renal_impairment=True):.1f} mg")
````

````python

# Reading line by line
print("Reading line by line:")
control_values = []
treatment_values = []

with open('sample_data.txt', 'r') as file:
    header = file.readline()  # Read the first line
    print(f"Header: {header.strip()}")
    
    # Read the rest of the lines
    for line in file:
        parts = line.strip().split('\t')
        if len(parts) == 3:
            sample_id, value, condition = parts
            value = float(value)
            
            if condition == "Control":
                control_values.append(value)
            elif condition == "Treatment":
                treatment_values.append(value)

# Calculate and print statistics
control_mean = sum(control_values) / len(control_values)
treatment_mean = sum(treatment_values) / len(treatment_values)

print(f"Control mean: {control_mean:.2f}")
print(f"Treatment mean: {treatment_mean:.2f}")
````

## 3. Working with CSV Data

CSV (Comma-Separated Values) files are commonly used to store tabular data in life sciences, from experimental results to clinical data.

````python
import csv

# Creating a CSV file with experimental data
with open('gene_expression.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Gene', 'Control_1', 'Control_2', 'Control_3', 'Treatment_1', 'Treatment_2', 'Treatment_3'])
    writer.writerow(['ACE2', 15.2, 14.8, 15.5, 42.3, 45.1, 39.7])
    writer.writerow(['IL6', 5.4, 4.9, 5.2, 22.8, 20.5, 25.1])
    writer.writerow(['TNF', 10.3, 11.2, 9.8, 13.5, 14.2, 12.9])
    writer.writerow(['VEGFA', 32.5, 30.8, 33.2, 25.4, 26.1, 24.8])
    writer.writerow(['TP53', 18.7, 19.2, 18.1, 17.9, 18.5, 18.3])
````

````python
# Reading CSV data
genes = []
control_means = []
treatment_means = []
fold_changes = []

with open('gene_expression.csv', 'r') as file:
    reader = csv.reader(file)
    header = next(reader)  # Skip the header
    
    for row in reader:
        gene = row[0]
        
        # Extract control and treatment values
        control_values = [float(x) for x in row[1:4]]
        treatment_values = [float(x) for x in row[4:7]]
        
        # Calculate means
        control_mean = sum(control_values) / len(control_values)
        treatment_mean = sum(treatment_values) / len(treatment_values)
        
        # Calculate fold change
        fold_change = treatment_mean / control_mean
        
        # Store the data
        genes.append(gene)
        control_means.append(control_mean)
        treatment_means.append(treatment_mean)
        fold_changes.append(fold_change)

# Display the results
print("Gene Expression Analysis:\n")
print(f"{'Gene':<6} {'Control':<10} {'Treatment':<10} {'Fold Change':<12} {'Status'}")
print("-" * 50)

for i in range(len(genes)):
    if fold_changes[i] > 1.5:
        status = "Up-regulated"
    elif fold_changes[i] < 0.67:  # 1/1.5
        status = "Down-regulated"
    else:
        status = "No change"
        
    print(f"{genes[i]:<6} {control_means[i]:<10.2f} {treatment_means[i]:<10.2f} {fold_changes[i]:<12.2f} {status}")
````

## 4. Introduction to NumPy

NumPy (Numerical Python) is a library for scientific computing in Python. It provides:
- Fast, efficient arrays (especially for large datasets)
- Mathematical functions for operations on these arrays
- Tools for working with linear algebra, Fourier transforms, and random numbers

````python
import numpy as npx
import numpy as np

# Creating NumPy arrays
# From lists
control = npx.array([23.5, 19.8, 18.2])
treatment = npx.array([45.1, 52.4])

print(f"Control array: {control}")
print(f"Treatment array: {treatment}")

# Basic array operations
print(f"\nBasic operations:")
print(f"Mean of control: {npx.mean(control):.2f}")
print(f"Standard deviation of treatment: {npx.std(treatment):.2f}")
print(f"Sum of all values: {npx.sum(npx.concatenate([control, treatment])):.2f}")
````

````python
# Creating arrays with specific patterns
# Array of zeros
zeros = np.zeros(5)
print(f"Array of zeros: {zeros}")

# Array of ones
ones = np.ones(5)
print(f"Array of ones: {ones}")

# Array with a range of values
range_array = np.arange(0, 10, 2)  # Start, stop, step
print(f"Range array: {range_array}")

# Linearly spaced values
linear_space = np.linspace(0, 1, 5)  # Start, stop, number of points
print(f"Linearly spaced: {linear_space}")

# Random arrays
random_values = np.random.random(5)  # 5 random values between 0 and 1
print(f"Random values: {random_values}")
````

````python
# Multidimensional arrays
# 2D array (matrix)
expression_data = np.array([
    [15.2, 14.8, 15.5, 42.3, 45.1, 39.7],  # ACE2
    [5.4, 4.9, 5.2, 22.8, 20.5, 25.1],     # IL6
    [10.3, 11.2, 9.8, 13.5, 14.2, 12.9],   # TNF
    [32.5, 30.8, 33.2, 25.4, 26.1, 24.8],  # VEGFA
    [18.7, 19.2, 18.1, 17.9, 18.5, 18.3]   # TP53
])

print(f"Shape of data: {expression_data.shape}")  # (rows, columns)

# Accessing elements and slices
print(f"First gene, first sample: {expression_data[0, 0]}")
print(f"Second gene, all samples: {expression_data[1, :]}")
print(f"All genes, treatment samples: {expression_data[:, 3:]}")

# Computing statistics
print(f"\nStatistics:")
print(f"Mean expression across all genes and conditions: {np.mean(expression_data):.2f}")

# Mean expression for each gene
gene_means = np.mean(expression_data, axis=1)  # axis=1 means operate across columns
print(f"Mean expression for each gene: {gene_means}")

# Mean for control vs treatment
control_means = np.mean(expression_data[:, 0:3], axis=1)
treatment_means = np.mean(expression_data[:, 3:6], axis=1)
print(f"\nControl means: {control_means}")
print(f"Treatment means: {treatment_means}")
````

````python
# Vectorized operations (much faster than loops)
fold_changes = treatment_means / control_means
log2_fold_changes = np.log2(fold_changes)

genes = ["ACE2", "IL6", "TNF", "VEGFA", "TP53"]
print("Gene Expression Analysis with NumPy:\n")
print(f"{'Gene':<6} {'Control':<10} {'Treatment':<10} {'Fold Change':<12} {'Log2 FC':<10}")
print("-" * 50)

for i in range(len(genes)):
    print(f"{genes[i]:<6} {control_means[i]:<10.2f} {treatment_means[i]:<10.2f} "
          f"{fold_changes[i]:<12.2f} {log2_fold_changes[i]:<10.2f}")
````
