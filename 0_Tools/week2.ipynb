# Control Flow in Python

In this notebook, we'll learn about control flow structures in Python, which allow us to control the execution of our code based on conditions and to repeat code multiple times. We'll cover:

- Boolean expressions
- Conditional statements (if, elif, else)
- Loops (for and while)
- Loop control (break, continue)

Let's get started!

## 1. Boolean Expressions

Boolean expressions evaluate to either `True` or `False`. They are fundamental for making decisions in your code.

````python
# Comparison operators
x = 10
y = 5

print(f"x = {x}, y = {y}")
print(f"x == y: {x == y}")  # Equal to
print(f"x != y: {x != y}")  # Not equal to
print(f"x > y: {x > y}")    # Greater than
print(f"x < y: {x < y}")    # Less than
print(f"x >= y: {x >= y}")  # Greater than or equal to
print(f"x <= y: {x <= y}")  # Less than or equal to
````

### Logical Operators

Python has three logical operators: `and`, `or`, and `not`.

````python
# Logical operators
a = True
b = False

print(f"a = {a}, b = {b}")
print(f"a and b: {a and b}")  # True if both are True
print(f"a or b: {a or b}")    # True if at least one is True
print(f"not a: {not a}")      # True if a is False
print(f"not b: {not b}")      # True if b is False

# Combined expressions
x = 10
y = 5
z = 20

print(f"x > y and x < z: {x > y and x < z}")  # True and True = True
print(f"x > y and x > z: {x > y and x > z}")  # True and False = False
print(f"x < y or x < z: {x < y or x < z}")    # False or True = True
print(f"not (x > y): {not (x > y)}")          # not True = False
````

### Truth Value Testing

In Python, many objects have a truth value that can be tested. The following values are considered False:
- `False`
- `None`
- `0` (zero)
- Empty sequences (`''`, `[]`, `()`)
- Empty dictionaries `{}`

Everything else is considered `True`.

````python
# Truth value testing
print(f"bool(False): {bool(False)}")
print(f"bool(None): {bool(None)}")
print(f"bool(0): {bool(0)}")
print(f"bool(''): {bool('')}")
print(f"bool([]): {bool([1,2,3])}")
#print(f"bool({}): {bool({})}")

print(f"bool(True): {bool(True)}")
print(f"bool(1): {bool(1)}")
print(f"bool('hello'): {bool('hello')}")
print(f"bool([1, 2, 3]): {bool([])}")
#print(f"bool({'a': 1}): {bool({'a': 1})}")
````

## 2. Conditional Statements

Conditional statements allow you to execute different code blocks based on conditions.

````python
# Simple if statement
age = 18

if age >= 18:
    print("You are an adult.")

# Note: Indentation is crucial in Python!
# The indented block is executed only if the condition is True.
````

````python
# if-else statement
age = 15

if age >= 18:
    print("You are an adult.")
else:
    print("You are not yet an adult.")
````

````python
# if-elif-else statement
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score}, Grade: {grade}")
````

````python
# Nested if statements
age = 20
has_license = True

if age >= 18:
    print("You are old enough to drive.")
    
    if has_license:
        print("You can drive.")
    else:
        print("You need to get a license first.")
else:
    print("You are not old enough to drive yet.")
````

### Conditional Expressions (Ternary Operator)

Python supports a compact way to write simple if-else statements.

````python
# Standard if-else
age = 20
if age >= 18:
    status = "adult"
else:
    status = "minor"
print(f"Status: {status}")

# Ternary operator (conditional expression)
age = 15
status = "adult" if age >= 18 else "minor"
print(f"Status: {status}")
````

## 3. For Loops

For loops are used to iterate over a sequence (like a list, tuple, string, or range) or other iterable objects.

````python
# For loop with range
print("Counting from 1 to 5:")
for i in range(1, 6):  # range(1, 6) generates numbers from 1 to 5
    print(i)

# range(start, stop, step)
print("\nEven numbers from 0 to 10:")
for i in range(0, 11, 2):  # Start at 0, go up to 10, step by 2
    print(i)
````

````python
# For loop with a list
fruits = ["apple", "banana", "cherry", "date"]

print("Fruit list:")
for fruit in fruits:
    print(f"- {fruit}")
````

````python
# For loop with a string
message = "Python"

print("Letters in 'Python':")
for char in message:
    print(char)
````

````python
# Loop with enumerate() - to get both index and value
fruits = ["apple", "banana", "cherry", "date"]

print("Fruit list with indices:")
for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")
````

````python
# Nested loops
pairs = []
for x in range(1, 4):  # 1, 2, 3
    for y in range(1, 3):  # 1, 2
        pairs.append((x, y))

print("All pairs:", pairs)
````

## 4. While Loops

While loops execute as long as a condition is True.

````python
# Basic while loop
count = 5

print("Countdown:")
while count > 0:
    print(count)
    count -= 1  # Decrement count by 1

print("Blast off!")
````

````python
# While loop with user input
# Note: In Jupyter, you might want to comment this out when not actively using it

password = "secret"
attempt = ""

while attempt != password:
    attempt = input("Enter the password: ")
    if attempt != password:
        print("Incorrect password, try again.")

print("Access granted!")
````

## 5. Loop Control

Python provides statements to control loop execution: `break` and `continue`.

````python
# break statement - exits the loop entirely
print("Looking for the magic number:")
for num in range(1, 10):
    print(f"Checking {num}...")
    if num == 7:
        print(f"Found the magic number: {num}!")
        break  # Exit the loop when the magic number is found

print("Search complete.")
````

````python
# continue statement - skips the rest of the current iteration
print("Processing only even numbers:")
for num in range(1, 10):
    if num % 2 != 0:  # If num is odd
        continue      # Skip the rest of this iteration
    print(f"Processing {num}...")
````

````python
# else clause with loops
# The else block executes after the loop completes normally (not via break)

# Example with for loop
print("Searching for a number divisible by both 5 and 7:")
for num in range(1, 36):  # Check numbers 1 through 35
    if num % 5 == 0 and num % 7 == 0:
        print(f"Found it! {num} is divisible by both 5 and 7.")
        break
else:  # This executes if no break occurred
    print("No number found that is divisible by both 5 and 7 in the given range.")
````

## 6. Putting It All Together

Let's combine what we've learned to create more complex examples.

````python
# Example 1: Finding prime numbers
def is_prime(n):
    """Check if a number is prime."""
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    
    # Check divisibility by numbers of the form 6k±1
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    
    return True

# Find all prime numbers between 1 and 50
primes = []
for num in range(1, 51):
    if is_prime(num):
        primes.append(num)

print(f"Prime numbers between 1 and 50: {primes}")
````

````python
# Example 2: Simple guessing game
import random

def play_guessing_game():
    """A simple number guessing game."""
    secret_number = random.randint(1, 10)
    attempts = 0
    max_attempts = 3
    
    print("I'm thinking of a number between 1 and 10.")
    
    while attempts < max_attempts:
        try:
            guess = int(input(f"Attempt {attempts + 1}/{max_attempts}: Enter your guess: "))
            attempts += 1
            
            if guess == secret_number:
                print(f"Congratulations! You guessed the number {secret_number} correctly!")
                break
            elif guess < secret_number:
                print("Too low! Try a higher number.")
            else:
                print("Too high! Try a lower number.")
                
        except ValueError:
            print("Please enter a valid number.")
            
    else:  # This executes if no break occurred (i.e., player didn't guess correctly)
        print(f"Game over! The secret number was {secret_number}.")

# Uncomment to play the game
# play_guessing_game()
````

## Bonus Challenge: Password Validator

Create a password validator that checks if a password is strong. A strong password must:
- Be at least 8 characters long
- Contain at least one uppercase letter
- Contain at least one lowercase letter
- Contain at least one digit
- Contain at least one special character (!@#$%^&*)

Your program should tell the user which requirements their password fails to meet.

````python
# Your solution here
password=input("Select a password:")

if len(password) <8:
    print("Be at least 8 characters long")

upper=False
for x in list(password):
    if x.isupper:
        upper=True
        break
if not upper :
    print("Contain at least one uppercase letter")

lower=False
for x in list(password):
    if x.islower:
        lower=True
        break
if not lower:
    print("Contain at least one lowercase letter")

digit=False
for x in list(password):
    for d in range(0,10):
        if str(d) == str(x):
            digit=True
            break
if not digit:
    print("Contain at least one digit")

special=False
for x in list(password):
    for s in list("!@#$%^&*"):
        if s==x:
            special=True
            break
if not special:
    print("Contain at least one special character (!@#$%^&*)")


if special and upper and digit and lower and len(password)>7:
    print("password is fine")
````
