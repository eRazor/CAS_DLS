## Implement perceptron performing logical OR gate

```python
import numpy as np
# define input
X = np.array( ((1,1), (1, 0), (0,1), (0,0)) )
print(X)
# set weights and biases
weights = np.array([1, 1])
threshold = -1.5
net = np.dot(X,weights)
print(net)

#compute and print output, where is bigger or equal 0 then 1 else 0
output = np.where(net + threshold >= 0, 1, 0)
print(output)
```

## Vizualize decision boundary for perceptron learning logical OR gate

```python
# hint: reuse the code from lecture ;)
# Let us implement the learning algorithm for the simple perceptron.

# We first import NumPy and specify the input array and the target array.
import numpy as np

# INPUT/TARGET ------------------------------------------------------------
# input
X = np.array( ((1,1), (1, 0), (0,1), (0,0)) )
# target
T = np.array([1,1,1,0])
X, T
# We set the initial weights and threshold/bias to zero.
# INITIAL WEIGHTS/BIAS -------------------------------------------
W_init = np.array([0,0])
b_init = 0
W_init, b_init
# We extend the input array and we include the bias in the weight matrix.
# EXTENDED INPUT ----------------------------------------------------------
W = np.append(W_init,b_init)
X_ext = np.hstack((X,np.ones((4,1))))
W, X_ext
```

Visualize decision boundary for perceptron learning XOR gate

```python
# hint: what is the input/output difference when implementing XOR compared to OR?
# PERCEPTRON LEARNING -------------------------------------------------------
epochs = 7 #number of epochs

for n in range(epochs):
    for i in range(len(X_ext)):
        y_i = np.where(np.dot(X_ext[i,:],W) >= 0, 1, 0)
        W = W + (T[i]-y_i)*X_ext[i,:]
    print(W)
y = np.where(np.dot(X_ext,W) >= 0, 1, 0)
#  Let us look at the target and the output and the weights found by the algorithm.
print(T)
print(y)
print(W)
```
