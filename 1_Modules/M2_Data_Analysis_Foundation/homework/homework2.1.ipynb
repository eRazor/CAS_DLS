## Understanding relative paths




Relative folder navigation using ```.``` and ```..``` is a way to refer to directories (folders) in relation to your current location in the file system. Here's how they work:

### 1. ```.``` (Single Dot)

Represents the current directory.
Useful when you want to run a command or reference a file in the current folder.

Example:

```./dataset1.csv``` 

points to the file dataset1.csv located in the same directory as your notebook.

### 2. ```..``` (Double Dot)

Represents the parent directory (one level up from the current directory).

Example:

```../data/nic-beach-water-quality-automated-sensors-1.csv``` 

points to a file located in the data folder that is located 1 level up from your homework folder

To load data from excel file with multiple tabs:


```Python
import pandas as pd

xls = pd.ExcelFile('path_to_file.xls')
df1 = pd.read_excel(xls, 'Sheet1')
df2 = pd.read_excel(xls, 'Sheet2')
```


To load only certain columns from Excel files and skip the first row

```Python
import pandas as pd

data = pd.read_excel("file.xlsx", parse_cols="A,B,C,K,L,M", skiprows=[0])
```
