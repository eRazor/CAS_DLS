# XML Demo

## Creating an xml output from the pirates csv file

```python
import pandas as pd
import xml.dom.minidom as md
import xml.etree.ElementTree as ET

# Read the CSV file
df = pd.read_csv('../data/pirates.csv')

# Create the root element
root = ET.Element('ships')

# Group by ShipID
ship_groups = df.groupby('ShipID')

# Create a ship element for each ShipID
for ship_id, pirates in ship_groups:
    # Create ship element with ShipID attribute
    ship = ET.SubElement(root, 'ship')
    ship.set('ShipID', str(ship_id))
    
    # Add pirates as subnodes
    for _, pirate_data in pirates.iterrows():
        pirate = ET.SubElement(ship, 'pirate')
        
        # Add all pirate attributes
        pirate.set('PirateID', str(pirate_data['PirateID']))
        pirate.set('FirstName', str(pirate_data['FirstName']))
        pirate.set('LastName', str(pirate_data['LastName']))
        pirate.set('Origin', str(pirate_data['Origin']))
        pirate.set('Teeth', str(pirate_data['Teeth']))
        pirate.set('Limbs', str(pirate_data['Limbs']))
        pirate.set('PositionID', str(pirate_data['PositionID']))

# Create the XML tree
tree = ET.ElementTree(root)

# Convert to string with pretty formatting
xml_str = ET.tostring(root, encoding='utf-8')
dom = md.parseString(xml_str)
pretty_xml = dom.toprettyxml(indent="  ")

# Save to file
with open('../data/pirates.xml', 'w+', encoding='utf-8') as f:
    f.write(pretty_xml)

# Display the first part of the XML
print(pretty_xml[:1000] + "...\n[XML output truncated]")
```

# Loading data from xml

```python

def load_pirate_xml(filepath):
    # Parse the XML file
    tree = ET.parse(filepath)
    root = tree.getroot()
    
    # Prepare data lists
    pirate_data = []
    
    # Iterate through ships and pirates
    for ship in root.findall('ship'):
        ship_id = ship.get('ShipID')
        
        # Get all pirates on this ship
        for pirate in ship.findall('pirate'):
            # Extract pirate attributes
            pirate_id = pirate.get('PirateID')
            first_name = pirate.get('FirstName')
            last_name = pirate.get('LastName')
            origin = pirate.get('Origin')
            teeth = pirate.get('Teeth')
            limbs = pirate.get('Limbs')
            
            # Append to data list
            pirate_data.append({
                'PirateID': pirate_id,
                'FirstName': first_name,
                'LastName': last_name,
                'Origin': origin,
                'Teeth': teeth,
                'Limbs': limbs,
                'ShipID': ship_id
            })
    
    # Create DataFrame
    pirates_df = pd.DataFrame(pirate_data)
    
    # Convert numeric columns to appropriate types
    pirates_df['PirateID'] = pd.to_numeric(pirates_df['PirateID'])
    pirates_df['Teeth'] = pd.to_numeric(pirates_df['Teeth'])
    pirates_df['Limbs'] = pd.to_numeric(pirates_df['Limbs'])
    pirates_df['ShipID'] = pd.to_numeric(pirates_df['ShipID'])
    
    return pirates_df
```

```python

df_pirates = load_pirate_xml('../data/pirates.xml')
```
